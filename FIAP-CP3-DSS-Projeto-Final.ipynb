Podemos ver Outliers nos campos numéricos `CreditScore`, `Age` e `NumOfProducts`. Agora nosso objetivo é examinar  esses outliers de forma mais detalhada. Para isso, utilizaremos uma função que calcula os limites inferior e superior com base no IQR (Interquartile Range). Essa função identifica os valores atípicos em relação à distribuição normal da variável analisada.

O processo envolve:

1. **Cálculo do IQR:** Diferença entre o terceiro e o primeiro quartil (Q3 - Q1).
2. **Definição dos limites:**
    - *Limite inferior:* Q1 - 1.5 * IQR
    - *Limite superior:* Q3 + 1.5 * IQR
3. **Identificação dos outliers:** Valores abaixo do limite inferior ou acima do limite superior são considerados outliers.

A função nos ajuda a listar esses outliers de forma prática para que possamos tomar decisões de tratamento mais assertivas.

```python
# Identifica Outliers
def outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  return outliers

# Exibe Outliers
def outliers_iqr_plot(df, outs):
  for i in range(len(outs)):
    out = outliers_iqr(df, outs[i])
    print(outs[i])
    display(out)
    # print(out.describe())
  pass

# Verificando outliers 
outs = ['CreditScore', 'Age', 'NumOfProducts']
outliers_iqr_plot(raw, outs)
```

Com essas três tabelas podemos ver quais linhas são outliers, sendo assim vamos consultá-las nas etapas de tratamento a seguir.

### Tratamento da coluna CreditScore
Nesta etapa, trabalharemos na coluna `CreditScore`, categorizando os clientes com base em faixas de pontuação para mitigar o impacto de outliers e garantir análises mais consistentes. Essas faixas seguem uma estrutura similar ao modelo de score utilizado por instituições financeiras no Brasil (como o Serasa), oferecendo uma forma de classificar o risco de crédito. A categorização adotada é a seguinte:

- **Excelente (4):** 701 a 1000 pontos, maior chance de conseguir crédito.
- **Bom (3):** 501 a 700 pontos, boa chance de conseguir crédito.
- **Baixo (2):** 301 a 500 pontos, chance média de obter crédito.
- **Muito baixo (1):** 0 a 300 pontos, baixa chance de obter crédito.

Ao transformar a variável em categorias numéricas, facilitamos a aplicação de algoritmos de machine learning que lidam melhor com dados ordenados. Isso também permite uma análise mais clara sobre a correlação entre o score e outras variáveis, como o risco de churn.

```python
# Função para categorizar em faixas de score com base nas faixas do Serasa
def score_range(score):
    if score >= 701:
        return 4
    elif 501 <= score <= 700:
        return 3
    elif 301 <= score <= 500:
        return 2
    elif 0 <= score <= 300:
        return 1
    else:
        return -1

# Criando a nova coluna com as faixas score
treated['CreditScoreRange'] = treated['CreditScore'].apply(score_range)
```

```python
sns.boxplot(data=treated, x='CreditScoreRange')
plt.title(f'Boxplot de CreditScoreRange', fontsize=14)
plt.tight_layout()
plt.show()
```

Após a categorização, os dados foram reorganizados em uma nova coluna chamada CreditScoreRange, agora representada por valores numéricos de 1 a 4, indicando a faixa correspondente. O boxplot gerado mostra a distribuição dessas faixas de forma clara e sem a presença de outliers, confirmando que a transformação foi bem-sucedida.

### Tratamento da coluna Age
Nesta etapa, trabalharemos com a coluna de idade (`Age`) para segmentar os clientes em faixas geracionais, facilitando a análise comportamental por grupos etários. Essa abordagem permite identificar padrões relevantes associados a diferentes gerações, como variações no comportamento financeiro ou na satisfação do cliente.

A categorização é feita utilizando o ano de nascimento, calculado subtraindo a idade dos clientes do ano de referência (2023). Com base no ano de nascimento, cada cliente é classificado em uma das seguintes faixas geracionais: 

| N | Geração              | Intervalo de Anos |
|---|----------------------|-------------------|
| 1 | Geração Alpha        | 2013 ~ 2025       |
| 2 | Geração Z            | 1997 – 2012       |
| 3 | Geração Millennial   | 1981 – 1996       |
| 4 | Geração X            | 1965 – 1980       |
| 5 | Baby Boomers         | 1946 – 1964       |
| 6 | Geração Silenciosa   | 1928 – 1945       |

Essas categorias são atribuídas como valores numéricos para facilitar a análise. A conversão torna a variável ordinal, permitindo seu uso em estatísticas descritivas e algoritmos de machine learning.

```python
year_reference = 2023

# Coluna com ano de nascimento
treated['YearOfBirth'] = year_reference - treated['Age']

# Função para categorizar em faixas geracionais com base no ano de nascimento
def generational_range(year):
    if year >= 2013:
        return 1
    elif 1997 <= year <= 2012:
        return 2
    elif 1981 <= year <= 1996:
        return 3
    elif 1965 <= year <= 1980:
        return 4
    elif 1946 <= year <= 1964:
        return 5
    else:
        return 6

# Criando a nova coluna com as faixas geracionais
treated['GenerationalRange'] = treated['YearOfBirth'].apply(generational_range)
```

```python
sns.boxplot(data=treated, x='GenerationalRange')
plt.title(f'Boxplot de GenerationalRange', fontsize=14)
plt.tight_layout()
plt.show()
```

O boxplot exibiu que ainda existem outliers associados à faixa etária mais alta, especificamente clientes da `Geração Silenciosa`. Para evitar que essa segmentação influencie negativamente a análise, realocamos esses clientes para a faixa dos `Baby Boomers`, por estarem próximos em idade. Assim, mantemos a representatividade da população mais idosa sem prejudicar a consistência dos dados.

```python
# Função para categorizar em faixas geracionais com base no ano de nascimento
def generational_range_fix(gen):
    if gen == 6:
        return 5
    else:
        return gen

# Criando a nova coluna com as faixas geracionais
treated['GenerationalRange'] = treated['GenerationalRange'].apply(generational_range_fix)
```

```python
sns.boxplot(data=treated, x='GenerationalRange')
plt.title(f'Boxplot de GenerationalRange', fontsize=14)
plt.tight_layout()
plt.show()
```

Com essa reorganização, o novo boxplot demonstra uma distribuição sem outliers, indicando que os dados estão mais equilibrados. Agora, os grupos etários podem ser analisados sem a interferência de valores extremos, permitindo estudos mais precisos sobre o impacto da geração nos padrões de comportamento dos clientes e facilitando a construção de modelos preditivos robustos.

### Tratamento da coluna Card Type
Estamos tratando a coluna `Card Type`, que contém categorias nominais dos tipos de cartão que os clientes utilizam. Essa coluna será convertida para um formato numérico para facilitar análises quantitativas, como correlações e classificações. Definimos uma hierarquia em que o cartão DIAMOND recebe o valor mais alto (4), representando o melhor nível, e o cartão SILVER recebe o menor valor (1), simbolizando a menor qualidade. Assim, transformamos essa variável categórica em uma ordinal, preservando a ordem de importância entre os cartões:

- **DIAMOND:** 4
- **PLATINUM:** 3
- **GOLD:** 2
- **SILVER:** 1

```python
# Dicionário para definir a hierarquia dos cartões
hierarquia = {'DIAMOND': 4, 'PLATINUM': 3, 'GOLD': 2, 'SILVER': 1}

# Adicionando uma nova coluna de rank com base na hierarquia
treated['Card Type Rank'] = treated['Card Type'].map(hierarquia)
treated.head(10)
```

Com a transformação aplicada, agora temos uma nova coluna chamada `Card Type Rank` que representa cada categoria de cartão de forma numérica, refletindo a hierarquia definida. Isso facilita a análise comparativa e o uso em modelos preditivos, pois elimina a necessidade de tratamento adicional das variáveis categóricas. Essa transformação também abre espaço para análises estatísticas, como correlações entre o tipo de cartão e a receita estimada (`Estimated Salary`), e para alimentar algoritmos de classificação, melhorando a performance dos modelos.

### Tratamento da coluna Geography
Para utilizar colunas categóricas nominais em modelos de Machine Learning, é necessário convertê-las em variáveis numéricas. Esse processo garante que os algoritmos compreendam as categorias sem atribuir ordens arbitrárias entre elas. No caso da coluna `Geography`, aplicamos o One-Hot Encoding, que cria uma nova coluna binária (0 ou 1) para cada categoria presente na variável original, garantindo que o modelo trate todas as categorias de forma justa e sem viés ordinal.

```python
# Aplicar One-Hot Encoding nas variáveis categóricas
dummies = pd.get_dummies(treated['Geography'], prefix='Geography').astype(int)

# Traz colunas para o DF principal sem perder a coluna 'Geography'
treated = pd.concat([treated, dummies], axis=1)
treated
```

Agora cada valor em `Geography` tem sua própria coluna, no caso quando formos usar essa informação no modelo traremos uma coluna a menos para evitar multicolinearidade.

# Etapa 4: Testes de Modelagem (Model)


Nesta etapa, implementamos dois modelos preditivos: um modelo de classificação para prever o churn (`Exited`) e um modelo de regressão para prever o número de produtos adquiridos (`NumOfProducts`). Avaliamos a performance utilizando logística e Random Forest para classificação e regressão linear e Random Forest Regressor para regressão. Também testamos a hipótese de que a normalização dos dados melhora a acurácia dos modelos. A seguir, apresentamos os resultados obtidos com análises detalhadas.

## Modelo de Classificação
**Objetivo:** Vamos utilizar modelos de aprendizado supervisionado para prever se um cliente irá sair da instituição (`Exited = 1`) ou permanecer (`Exited = 0`). O foco é construir um modelo robusto que possa antecipar comportamentos de churn com base nas características do cliente e melhorar a retenção.

### Preparando Base
Selecionamos as colunas mais relevantes para a modelagem, priorizando variáveis que apresentaram maior impacto em análises anteriores:

- **CreditScoreRange:** Faixa de pontuação de crédito.
- **Age:** Idade do cliente.
- **Balance:** Saldo na conta.
- **EstimatedSalary:** Salário estimado.
- **IsActiveMember:** Status do cliente (ativo/inativo).
- **Geography_Spain e Geography_Germany:** Variáveis dummies da localização.
- **Card Type Rank:** Rank do tipo de cartão utilizado.

```python
# Definir variáveis independentes (X) e dependente (y)
X = treated[['CreditScoreRange', 'Age', 'Balance', 'EstimatedSalary', 'IsActiveMember', 'Geography_Spain', 'Geography_Germany', 'Card Type Rank']]
y = treated['Exited']

# Separar a base em 70% treino e 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Dividimos a base em 70% para treino e 30% para teste. Essa abordagem garante que o modelo possa aprender com uma amostra significativa dos dados e, posteriormente, seja avaliado em um conjunto não visto anteriormente, evitando overfitting. O parâmetro `random_state=42` assegura que a divisão seja sempre a mesma, permitindo reprodutibilidade dos resultados.

### Criando Modelo
Nesta etapa, implementamos dois modelos supervisionados: Regressão Logística e Random Forest Classifier, com o objetivo de prever o churn (clientes que saem da instituição). A seguir, apresentamos o processo de criação, predições e avaliação dos modelos com acurácia e o relatório de classificação.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Criando e treinando modelos
logistic = LogisticRegression().fit(X_train, y_train)
rf = RandomForestClassifier(random_state=42).fit(X_train, y_train)

# Predições e avaliação
y_pred_log = logistic.predict(X_test)
y_pred_rf = rf.predict(X_test)

# Avaliando acurácia
print("Acurácia - Regressão Logística:", accuracy_score(y_test, y_pred_log))
print("Acurácia - Random Forest:", accuracy_score(y_test, y_pred_rf))

# Relatório de Classificação
print("\nRelatório de Classificação - Random Forest")
print(classification_report(y_test, y_pred_rf))
```

O modelo de Random Forest apresentou uma acurácia superior (82.16%) em relação à Regressão Logística (79.44%). Além disso, o Random Forest também obteve um melhor f1-score para o churn (41%), embora ambos os modelos tenham tido dificuldades em identificar corretamente os clientes que saíram. Isso é evidente pela baixa precisão e recall para a classe "Sim (1)", indicando que mais variáveis e talvez uma engenharia de atributos adicional podem ser necessárias para melhorar o desempenho.

### Teste de Hipótese: Normalizar os campos numéricos melhora a acurácia do modelo?
A normalização coloca todas as variáveis na mesma escala, beneficiando algoritmos que são sensíveis à magnitude dos dados, como a Regressão Logística. Vamos verificar se a normalização melhora o desempenho tanto da Regressão Logística quanto do Random Forest Classifier.

```python
from sklearn.preprocessing import StandardScaler

# Separar dados numéricos e normalizar
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[['CreditScoreRange', 'Age', 'Balance', 'EstimatedSalary']])
X_test_scaled = scaler.transform(X_test[['CreditScoreRange', 'Age', 'Balance', 'EstimatedSalary']])

# Substituir as colunas normalizadas nas bases originais
X_train_class = X_train.copy()
X_test_class = X_test.copy()

X_train_class[['CreditScoreRange', 'Age', 'Balance', 'EstimatedSalary']] = X_train_scaled
X_test_class[['CreditScoreRange', 'Age', 'Balance', 'EstimatedSalary']] = X_test_scaled

# Treinar novamente a Regressão Logística e Random Forest com dados normalizados
logistic_scaled = LogisticRegression(random_state=42).fit(X_train_class, y_train)
rf_scaled = RandomForestClassifier(random_state=42).fit(X_train_class, y_train)

# Avaliação dos modelos normalizados
y_pred_log_scaled = logistic_scaled.predict(X_test_class)
y_pred_rf_scaled = rf_scaled.predict(X_test_class)

print("Acurácia - Regressão Logística (Normalizada):", accuracy_score(y_test, y_pred_log_scaled))
print("Acurácia - Random Forest Classifier (Normalizada):", accuracy_score(y_test, y_pred_rf_scaled))

print("\nRelatório de Classificação - Regressão Logística (Normalizada):")
print(classification_report(y_test, y_pred_log_scaled))
```

A normalização aumentou a acurácia da Regressão Logística de 79.44% para 80.68%, indicando que este algoritmo foi beneficiado pela normalização dos campos numéricos. Por outro lado, o Random Forest manteve praticamente a mesma performance, com 82.18%, demonstrando que modelos baseados em árvores não são significativamente afetados pela escala das variáveis.



## Modelo de Regressão
**Objetivo:** Prever o número de produtos adquiridos por um cliente (`NumOfProducts`) utilizando técnicas de regressão supervisionada. A análise nos permitirá identificar quais variáveis influenciam mais diretamente na quantidade de produtos adquiridos, auxiliando em estratégias de recomendação e vendas.



### Preparando Base
Vamos utilizar as seguintes colunas:

- **Age:** Idade do cliente.
- **CreditScoreRange:** Faixa de pontuação de crédito.
- **Balance:** Saldo do cliente.
- **IsActiveMember:** Status ativo/inativo.
- **EstimatedSalary:** Salário estimado.
- **Card Type Rank:** Rank do cartão de crédito.

```python
# Definindo X e y para regressão
X_reg = treated[['Age', 'CreditScore', 'Balance', 'IsActiveMember', 'EstimatedSalary', 'Card Type Rank']]
y_reg = treated['NumOfProducts']

# Separar dados em treino e teste
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)
```

Aqui, definimos as variáveis independentes (X) e a variável dependente (y). Em seguida, utilizamos a função train_test_split para dividir os dados, onde 70% dos dados são reservados para treinamento e 30% para teste, garantindo uma amostra não vista para avaliar o desempenho do modelo. O parâmetro `random_state=42` assegura que a divisão seja reproduzível.



### Criando Modelo
Implementamos dois algoritmos de regressão supervisionada: Regressão Linear e Random Forest Regressor, com o objetivo de comparar seu desempenho na previsão do número de produtos adquiridos (NumOfProducts).



```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Treinar modelos
linear_model = LinearRegression().fit(X_train_reg, y_train_reg)
rf_model = RandomForestRegressor(random_state=42).fit(X_train_reg, y_train_reg)

# Fazer predições
y_pred_lin = linear_model.predict(X_test_reg)
y_pred_rf = rf_model.predict(X_test_reg)

# Avaliar desempenho
print("MSE - Regressão Linear:", mean_squared_error(y_test_reg, y_pred_lin))
print("MSE - Random Forest Regressor:", mean_squared_error(y_test_reg, y_pred_rf))
```

A Regressão Linear apresentou um MSE ligeiramente menor que o Random Forest Regressor, sugerindo uma performance levemente superior. No entanto, ambos os modelos não se mostraram ideais para prever com precisão o número de produtos adquiridos. Isso indica que as variáveis selecionadas não possuem uma relação forte com a quantidade de produtos, prejudicando a capacidade dos modelos em generalizar bem. Essa limitação pode ser explicada pelo fato de o dataset ter sido originalmente estruturado para análise de churn e não para previsão do número de produtos. Variáveis mais diretamente relacionadas ao comportamento de aquisição de produtos poderiam melhorar a performance do modelo.

### Teste de Hipótese: Normalizar os campos numéricos melhora a acurácia do modelo?
Verificamos se a normalização dos campos numéricos pode melhorar a precisão dos modelos de regressão utilizados para prever o número de produtos adquiridos. A normalização pode beneficiar modelos lineares, como a Regressão Linear, que são sensíveis à magnitude dos dados.



```python
# Normalizando os dados de treinamento e teste para regressão
X_train_reg_scaled = scaler.fit_transform(X_train_reg[['Age', 'CreditScore', 'Balance', 'EstimatedSalary']])
X_test_reg_scaled = scaler.transform(X_test_reg[['Age', 'CreditScore', 'Balance', 'EstimatedSalary']])

# Substituir as colunas normalizadas nas bases originais
X_train_reg_norm = X_train_reg.copy()
X_test_reg_norm = X_test_reg.copy()

X_train_reg_norm[['Age', 'CreditScore', 'Balance', 'EstimatedSalary']] = X_train_reg_scaled
X_test_reg_norm[['Age', 'CreditScore', 'Balance', 'EstimatedSalary']] = X_test_reg_scaled

# Treinando Regressão Linear e Random Forest Regressor com dados normalizados
linear_model_scaled = LinearRegression().fit(X_train_reg_norm, y_train_reg)
rf_model_scaled = RandomForestRegressor(random_state=42).fit(X_train_reg_norm, y_train_reg)

# Avaliação dos modelos
y_pred_lin_scaled = linear_model_scaled.predict(X_test_reg_norm)
y_pred_rf_scaled = rf_model_scaled.predict(X_test_reg_norm)

print("MSE - Regressão Linear (Normalizada):", mean_squared_error(y_test_reg, y_pred_lin_scaled))
print("MSE - Random Forest Regressor (Normalizada):", mean_squared_error(y_test_reg, y_pred_rf_scaled))
```

A normalização dos dados não trouxe melhorias significativas para os modelos de regressão. O MSE da Regressão Linear permaneceu inalterado após a normalização, indicando que este modelo já estava performando no limite com os dados disponíveis. Por outro lado, o Random Forest Regressor apresentou uma melhora marginal de 0,01%, o que reforça que modelos baseados em árvores não são afetados pela escala dos dados.
